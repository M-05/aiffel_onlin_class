```python
import face_recognition
import os

def get_cropped_face(image_file):
    image = face_recognition.load_image_file(image_file)
    face_locations = face_recognition.face_locations(image)
    a, b, c, d = face_locations[0]
    cropped_face = image[a:c, d:b, :]
    
    return cropped_face
```

`face_recognition.face_encodings`(
    face_image,
    known_face_locations=None,
    __num_jitters=1__,
    model='small',
)  
Given an image, return the __128-dimension__ face encoding for each face in the image.

```python
image_file = os.path.join(dir_path, 'obama.jpg')
face = get_cropped_face(image_file)   # 얼굴 영역을 구하는 함수(이전 스텝에서 구현)

# 얼굴 영역을 가지고 얼굴 임베딩 벡터를 구하는 함수
def get_face_embedding(face):
    return face_recognition.face_encodings(face)

embedding = get_face_embedding(face)  
len(embedding)
# array([-0.08126248,  0.11014761, -0.01082448, -0.05268792,  0.01033717,
# 'list' object has no attribute 'shape'
```

`os.path.splitext`(p)  
Split the extension from a pathname.

```python
def get_face_embedding_dict(dir_path):
    file_list = os.listdir(dir_path)
    embedding_dict = {}
    
    for file in file_list:
        img_path = os.path.join(dir_path, file)
        face = get_cropped_face(img_path)
        embedding = get_face_embedding(face)
        
# 얼굴영역 face가 제대로 detect되지 않으면  len(embedding)==0인 경우가 발생하므로 
        if len(embedding) > 0:  
            embedding_dict[os.path.splitext(file)[0]] = embedding[0]
# os.path.splitext(file)[0]에는 이미지파일명에서 확장자를 제거한 이름이 담깁니다. 
        
    return embedding_dict
```

Matrix or vector norm.  

`ord` {non-zero int, inf, -inf, 'fro', 'nuc'}, optional  
    Order of the norm (see table under ``Notes``).  
    inf means numpy's
    `inf` object.   
    The default is None.

### Step3. 얼굴 영역의 임베딩 추출하기

```python
# 얼굴 영역을 가지고 얼굴 임베딩 벡터를 구하는 함수
def get_face_embedding(face):
    return face_recognition.face_encodings(face)
```

```python
embedding = get_face_embedding(face1)
embedding
```

```python
def get_face_embedding_dict(dir_path):
    file_list = os.listdir(dir_path)
    embedding_dict = {}
    
    for file in file_list:
        img_path = os.path.join(dir_path, file)
        face = get_cropped_face(img_path)
        embedding = get_face_embedding(face)
        if len(embedding) > 0:  
            embedding_dict[os.path.splitext(file)[0]] = embedding[0]
        
    return embedding_dict
```

```python
embedding_dict = get_face_embedding_dict(dir_path)
```

```python
embedding_dict["lee2"]
```

```python
def get_distance(name1, name2):
    return np.linalg.norm(embedding_dict[name1] - embedding_dict[name2], ord= 2)
```

```python
get_distance('lee1', 'lee2')
```

```python
get_distance('lee1', 'kimjongun')
```

```python
get_distance('lee2', 'sin')
```

```python
get_distance('kimjongun', 'yun')
```

```python
def get_sort_key_func(name1):
    def get_distance_from_name1(name2):
        return get_distance(name1, name2)
    return get_distance_from_name1
```

### Step4. 모은 연예인들과 비교하기
get_nearest_face(name, top=5)

```python
def get_nearest_face(name, top=7):
    sort_key_func = get_sort_key_func(name)
    sorted_faces = sorted(embedding_dict.items(), key=lambda x:sort_key_func(x[0]))
    
    for i in range(top + 1):
        if i == 0 :   # 첫번째로 나오는 이름은 자기 자신일 것이므로 제외합시다. 
            continue
        if sorted_faces[i]:
            print('순위 {} : 이름({}), 거리({})'.format(i, sorted_faces[i][0], sort_key_func(sorted_faces[i][0])))
```
