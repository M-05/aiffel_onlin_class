### load data 

```python
df_daily = pd.read_csv("daily_train.csv")
df_immo = pd.read_csv("immoral.csv")
df_dktc = pd.read_csv("merged_dktc_train.csv")
df_exp = pd.read_csv("dktc_ori_exp_0422_82e-2.csv")
df_imp = pd.read_csv("dktc_ori_imp_0422_82e-2.csv")
```

```python
df_immo['Unnamed: 0'] = df_immo['class']
df_immo = df_immo[['Unnamed: 0', 'conversation']]
df_immo.rename(columns={"Unnamed: 0" : "class"}, inplace=True)
```

```python
df_dktc = df_dktc[['class', "conversation"]]
```

```python
df_imp = df_imp[['class', "conversation"]]
df_exp = df_exp[['class', "conversation"]]
```

### dktc data

```python
df_dktc['class'].value_counts()
```

```python
df_dktc1 = df_dktc[df_dktc['class'] == "기타 괴롭힘 대화"].iloc[:10][::2]
df_dktc2 = df_dktc[df_dktc['class'] == "갈취 대화"].iloc[:10][::2]
df_dktc3 = df_dktc[df_dktc['class'] == "직장 내 괴롭힘 대화"].iloc[:10][::2]
df_dktc4 = df_dktc[df_dktc['class'] == "협박 대화"].iloc[:10][::2]
```

### explicit & implicit

```python
df_exp1 = df_exp[df_exp['class'] == "기타 괴롭힘 대화"].iloc[:5]
df_exp2 = df_exp[df_exp['class'] == "갈취 대화"].iloc[:5]
df_exp3 = df_exp[df_exp['class'] == "직장 내 괴롭힘 대화"].iloc[:5]
df_exp4 = df_exp[df_exp['class'] == "협박 대화"].iloc[:5]
```

```python
df_imp1 = df_imp[df_imp['class'] == "기타 괴롭힘 대화"].iloc[:5]
df_imp2 = df_imp[df_imp['class'] == "갈취 대화"].iloc[:5]
df_imp3 = df_imp[df_imp['class'] == "직장 내 괴롭힘 대화"].iloc[:5]
df_imp4 = df_imp[df_imp['class'] == "협박 대화"].iloc[:5]
```

### concatenate

```python
concat_df = pd.concat([df_daily.iloc[:5], df_immo.iloc[:5], 
                        df_dktc1, df_dktc2, df_dktc3, df_dktc4], 
                       join='inner', axis=0)
concat_df
```

```python
concat_exp = pd.concat([df_exp1, df_exp2, df_exp3, df_exp4], 
                       join='inner', axis=0)
concat_exp
```

```python
concat_imp = pd.concat([df_imp1, df_imp2, df_imp3, df_imp4], 
                       join='inner', axis=0)
concat_imp
```

### shuffle

```python
for i in range(10):
    concat_df = concat_df.sample(frac=1)
    concat_exp = concat_exp.sample(frac=1)
    concat_imp = concat_imp.sample(frac=1)
```

### wandb

```python
wandb.init(project="model_1_0423")
```

```python
wandb.run.log({"Team GAL's Total Dataset" : wandb.Table(dataframe=concat_df)})
wandb.run.log({"Explicit Dataset" : wandb.Table(dataframe=concat_exp)})
wandb.run.log({"Implicit dataset" : wandb.Table(dataframe=concat_imp)})
```

```python
wandb.finish()
```
