# 가위바위보 이미지 개수 총합에 주의하세요.

```python
import numpy as np

def load_data(img_path, number_of_data=300): 
    # 가위 : 0, 바위 : 1, 보 : 2
    img_size=28
    color=3
    #이미지 데이터와 라벨(가위 : 0, 바위 : 1, 보 : 2) 데이터를 담을 행렬(matrix) 영역을 생성합니다.
    imgs= np.zeros(number_of_data * img_size * img_size * color, dtype=np.int32).reshape(number_of_data, 
                                                                                  img_size,img_size, 
                                                                                  color)
    labels= np.zeros(number_of_data, dtype= np.int32)

    idx=0
#scissor data
    for file in glob.iglob(img_path+'/scissor/*.jpg'):
        img = np.array(Image.open(file), dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=0   # 가위 : 0
        idx=idx+1
#rock data
    for file in glob.iglob(img_path+'/rock/*.jpg'): 
        img = np.array(Image.open(file), dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=1   # 바위 : 1
        idx=idx+1  
#paper data    
    for file in glob.iglob(img_path+'/paper/*.jpg'):
        img = np.array(Image.open(file),dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=2   # 보 : 2
        idx=idx+1
        
    print("학습데이터(x_train)의 이미지 개수는", idx,"입니다.")
    return imgs, labels

image_dir_path = os.getenv("HOME") + "/aiffel/rock_scissor_paper"
(x_train, y_train)= load_data(image_dir_path)
x_train_norm = x_train / 255.0   
# 입력은 0~1 사이의 값으로 정규화

print("x_train shape: {}".format(x_train.shape))
print("y_train shape: {}".format(y_train.shape))
```

```python
def load_data(img_path, number_of_data= 1350): # 300 -> 1500 -> 1350
    # 가위 : 0, 바위 : 1, 보 : 2
    img_size= 64 # 28 -> 256 -> 128 -> 64
    color=3
    #이미지 데이터와 라벨(가위 : 0, 바위 : 1, 보 : 2) 데이터를 담을 행렬(matrix) 영역을 생성합니다.
    imgs= np.zeros(number_of_data * img_size * img_size * color, dtype=np.int32).reshape(number_of_data, 
                                                                                  img_size,img_size, 
                                                                                  color)
    labels= np.zeros(number_of_data, dtype= np.int32)

    idx=0
#scissor data
    for file in glob.iglob(img_path+'/scissor/*.jpg'):
        img = np.array(Image.open(file), dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=0   # 가위 : 0
        idx=idx+1
#rock data
    for file in glob.iglob(img_path+'/rock/*.jpg'): 
        img = np.array(Image.open(file), dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=1   # 바위 : 1
        idx=idx+1  
#paper data    
    for file in glob.iglob(img_path+'/paper/*.jpg'):
        img = np.array(Image.open(file),dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=2   # 보 : 2
        idx=idx+1
        
    print("학습데이터(x_full_data)의 이미지 개수는", idx,"입니다.")
    return imgs, labels
```

 첫 번째에서는 강압적으로 epoch수를 500회를 늘림으로써 100%의 정확도를 찍었습니다.(의미 없죠 본인꺼랑 비교한느디...)  
그러한 과정 중 가위 바위 보 (0 1 2)의 레이블인 정수 값을 찾는 문제이기에  
- loss를 sparse_categorical_crossentropy 사용했고  
- optimizer는 adam 믿어봐달라는 절실한 갓진표님의 의견을 따라서 사용했습니다.  
- 정확도 구분하는 프로젝트다 보니 metrics 부분을 accuracy를 사용했습니다.  
  
어느 새벽이 밝았습니다.  
결국 1350장의 사진을 모으고 이름도 정리(__batchnamer__ 어플 -> `수경님` 추천)하고  
업로드하는 부분이 epoch을 기다리는 것처럼 밀당을 잘 하더라구요.
그렇게 이미지가 준비가 되고 나서 split하는 부분에서 많은 시간을 할애했습니다.  
애매하게 아는 부분이기에 여기저기서 자료를 확인하고 前프로젝트 등을 확인하여 대에에충 이런 방식으로 분리하는 방법을 알게되었습니다.  

저의 큰 과제인  VGG 16 모델 구성... 실패...  
알고 보니 이미지 사이즈랑 어울리는 layer를 쌓으지 못했던 부분이여서  
(28, 28) 4-layer -> (256, 256) 16-layer -> (128, 128) 9-layer -> (64, 64) 6-layer로 변하는 과정을 거쳤습니다.    
현재로써는 명확하게는 잘 모르겠지만  
total params 계산법도 힘겹게 혼자만의 싸움에서 이겨내서 알게된 상태에서  
지금쯤이면 다른 사이트에 설명이 납득이 되는 부분이 늘었을 것이라 저는 생각합니다.  

1. sigmoid -> softmax  
2. learning rate 0.0000001 - 결국, -> 0.001  
3. layer 안에는 batchnormal & dropout은 빼는 부분이 더 좋은 정확도가 나왔습니다.    
4. 커널이 죽는 문제가 또 발생하여서 이미지 사이즈를 변경하여 결국 64, 64의 형태가 되었습니다.  
등등의 과정을 거쳐 __accuracy: 0.9500__라는 값이 나왔지 않았을까 생각합니다.  

 아침이 밝았네요...  
하지만 저는 train test 데이터를 제대로 분리를 하지 못했더라구요..  
`'갓 성돈'`님의 힌트로 알게 되었습니다.  
>ex) 100장의 저의 손 사진를 분리  
대충, 80장 -> train, 20장 -> test로 분리  
80장의 제 손 이미지로 - - -> 남은 20장의 제 손을 체크하니  
정확도가 남들보다 빨리 올라갈 수 밖에 없는 구조네요.  
결국, '100장의 제 손 이미지는 1개의 데이터로 생각해야한다'는 부분을 깨닫게 되었네요.

다시 원점으로 와서 이미지 정제?하는 과정을 거쳐서

이번을 계기로 이제서야 드디오 train_test_split라는 의미를 0.000001%는 얻어갔네요.
