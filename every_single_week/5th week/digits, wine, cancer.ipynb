###  import

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report as ccr
import pandas as pd
```

###  

```python
load_digits = load_digits()
```

```python
load_digits.keys()
```

```python
digits_data = load_digits.data # 
digits_label = load_digits.target #
```

```python
digits_data.shape, digits_label.shape, load_digits.images.shape
```

```python
len(load_digits.feature_names)
```

```python
load_digits.target_names.shape
```

```python
digits_target_names = load_digits.target_names
digits_target_names
```

```python
digits_df = pd.DataFrame(data = digits_data, columns = load_digits.feature_names)
digits_df
```

```python
digits_df.describe()[:]
```

### train, test  

```python
X_train, X_test, y_train, y_test = train_test_split(digits_data, digits_label,
                                                    test_size = 0.25,
                                                    random_state = 5)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

###   

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state = 5)
```

```python
dtc.fit(X_train, y_train)
y_pred1 = dtc.predict(X_test)
print(ccr(y_test, y_pred1, zero_division = 0))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = DecisionTreeClassifier(random_state = 5)
```

```python
rfc.fit(X_train, y_train)
y_pred2 = rfc.predict(X_test)
print(ccr(y_test, y_pred2, zero_division = 0))
```

```python
sgdc.fit(X_train, y_train)
y_pred4 = sgdc.predict(X_test)
print(ccr(y_test, y_pred4, zero_division = 0))
```
