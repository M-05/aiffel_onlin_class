### 데이터 분석

```python
temp_var = sales_data
temp_var.tail(8 +1)
```

temp_var변수에 date라는 열을 생성시키고,  
sales_data변수 안에 있는 결제일시의 내용들을 전부 7번째까지 부분만 뽑아서  
temp_var변수에 넣으세요(할당하세요)

```python
temp_var['date'] = sales_data["결제일시"].str[:7]
temp_var
```

Dtype 확인

```python
temp_var.info()
```

str.replace는 일회성이기에  
다시 해당 변수 or 변수의 해당 열에 할당 시켜야 적용된다.

```python
temp_var["결제금액"] = sales_data["결제금액"].str.replace(",","")
temp_var
```

```python
temp_var["결제금액"] = temp_var["결제금액"].astype('int64')
```

```python
temp_var.info()
```

temp_var의 payment열을 date 기준으로 집계하기  
-> temp_var변수 안에 payment열을 만들고 temp_var의 date 기준으로 총합을 구해라.  
 집계 集計 total aggregate 총액 합계. sum
 
-> 질문 자체가 잘못 됨.  
payment안에 결제금액 수치를 대입하고 date를 그 옆에 열에 붙여서 비교할 수 있게 하라.

```python
temp_var["payment"] = temp_var["결제금액"]
temp_var
```

변수에 할당 시켜야 저장이 되고 다양하게 이용 가능

```python
total = temp_var["payment"].groupby(temp_var["date"]).sum()
total
```

```python
total.plot(kind = 'line')

plt.title("a Sales Statement")
plt.xlabel('Month')
plt.ylabel('amount')
plt.show()
```
