## 18 - 3 시각화

```python
import matplotlib.pyplot as plt
```

```python
cus_sam = cus_sam.sort_values(by = ['HP'], ascending = False)
cus_sam = cus_sam.reset_index(drop = True)
```

```python
#line
plt.plot(cus_sam['HP'])
plt.show()

# 동시 추출 불가능인듯...
```

```python
#bar
plt.bar(cus_sam.index, cus_sam['HP'])
plt.show()
```

```python
#histogram
plt.hist(cus_sam['HP'])
plt.show()
```

```python
# Box
plt.boxplot(cus_sam["HP"])
plt.show()
```

```python
#scatter
plt.scatter(cus_sam['HP'], cus_sam['Defense'])
```

```python
#pie
pie_d = cus_sam.groupby(["MobileCompany"]).count().Name

plt.pie(pie_d, labels = pie_d.index, autopct = '%1.1f%%')
plt.show()
```
